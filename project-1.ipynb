```python
import pandas as pd
```

```python
file = "flightdata5y.csv"
```

```python
df = pd.read_csv(file)
df.head()
```

```python
df.shape
```

```python
airlines = ["AA", "UA","DL","WN","NK"]
airline_info = df[df.AIRLINECODE.isin(airlines)]
airline_info.head()
```

```python
airline_info.shape
```

```python
airports = ["ATL", "LAX","CHI","DFW","DEN", "JFK"]

data = airline_info.loc[airline_info.ORIGINAIRPORTCODE.isin(airports) | airline_info.DESTAIRPORTCODE.isin(airports)]


data.head()
```

```python
data.shape
```

```python
data.dropna(how="any")
```

```python
data.dtypes
```

```python
delay_data = data.loc[(data["DEPDELAY"] > 0) | (data["ARRDELAY"] > 0) | (data["CANCELLED"] == "T")]

delay_data.head()
```

```python
delay_data.shape
```

```python
airlinecode = data.groupby(["AIRLINECODE"])
airlinecode.count().head()
```

```python
data["AIRLINECODE"].value_counts()
```

```python

# Using GroupBy in order to separate the data into fields according to "state" values
ordered_year = data.groupby(["Year"])

# The object returned is a "GroupBy" object and cannot be viewed normally...
#print(grouped_usa_df)

# In order to be visualized, a data function must be used...
ordered_year.count().head(10)
```

```python
ordered_year_month = data.groupby(["Year", "MONTH"])
ordered_year_month.count().head(30)
```

```python
grouped_ymcode = data.groupby(["Year", "MONTH","AIRLINECODE"])
grouped_ymcode.count().head(20)
```
